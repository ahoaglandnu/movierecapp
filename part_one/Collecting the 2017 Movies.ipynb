```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
r = requests.get('https://en.wikipedia.org/wiki/2017_in_film').text
```

```python
soup = BeautifulSoup(r, 'html.parser')
```

```python
print(soup.prettify)
```

```python
rows = []
for table in soup.findAll('table', {'class': 'wikitable'}):
    for row in table.findAll('tr')[1:]:
        cells = []
        for cell in row.findAll('i'):
            cells.append(cell.text)
        rows.append(cells)
```

```python
for row in rows:
    print(row)
```

```python
print(len(rows))
print(rows)
```

```python
# manual slice count
del rows[:34]
print(rows)
```

```python
clean_rows = [l[0] for l in rows]
```

```python
clean_rows
```

```python
len(clean_rows)
```

```python
movie_id = range(0,218)
```

```python
df = pd.DataFrame({'movie_id': movie_id,'movies': clean_rows})
```
